Searching for sampling methods and plotting the keyword and publication trend of the results

```python
topics = ['monte carlo', 'mc sampling', 'latin hypercube', 'lhs', 'morris screening', 'morris method']
sampling = wosis.keyword_matches(RC, ['monte carlo', 'mc sampling', 'latin hypercube', 'lhs', 'morris screening', 
                                      'morris method'], threshold=95.0)
```

```python
all_samp = sampling.combine_recs()
```

```python
wos_plot.plot_kw_trend(all_samp, title='Sampling Keyword Trend')
```

```python
wos_plot.plot_pub_trend(all_samp, title='Sampling Publication Trend')
```
